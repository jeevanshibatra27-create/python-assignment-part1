```python
# Original essay submitted by student
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Step 1: Remove extra spaces at the start and end
clean_essay = essay.strip()
print("Step 1 - Clean Essay:")
print(clean_essay, "\n")

# Step 2: Turn it into Title Case (every word starts with capital letter)
title_case_essay = clean_essay.title()
print("Step 2 - Title Case:")
print(title_case_essay, "\n")

# Step 3: Count how many times 'python' appears
# We'll ignore the case, so converting everything to lowercase first
python_count = clean_essay.lower().count("python")
print("Step 3 - How many times 'python' appears:")
print(python_count, "\n")

# Step 4: Replace 'python' with 'Python 🐍'
# Using the cleaned essay so we catch all instances
replaced_essay = clean_essay.replace("python", "Python 🐍")
print("Step 4 - Replace 'python' with 'Python 🐍':")
print(replaced_essay, "\n")

# Step 5: Split essay into sentences
sentences = clean_essay.split(". ")
print("Step 5 - Sentences List:")
print(sentences, "\n")

# Step 6: Print each sentence nicely, numbered, and make sure it ends with a dot
print("Step 6 - Numbered Sentences:")
for i, sentence in enumerate(sentences, start=1):
    # Add dot if it’s missing
    if not sentence.endswith("."):
        sentence += "."
    print(f"{i}. {sentence}")
```
